```python

import pandas as pd

df = pd.read_csv('./files/sales_data_sample.csv', encoding='latin1')
# Get the first 5 rows
df.head()
```

```python
# Describe df 
df.describe()
```

```python
# Get the number of rows that do not contain valid values in the numerical columns
# in df 
# Count NaN or missing values in each column
invalid_counts = df.isna().sum()

# Convert non-numeric values to NaN and then count them in each column
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
invalid_counts += df.isna().sum()

# print(invalid_counts)
# Print only values that are greater than zero
# print(invalid_counts[invalid_counts > 0])

# Print invalid counts where the value is greater than zero and not equal to 5646
print(invalid_counts[(invalid_counts > 0) & (invalid_counts != 5646)])
```
