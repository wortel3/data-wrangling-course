```python
import tabula

# Read pdf into a list of DataFrame
pdf  = tabula.read_pdf("foo.pdf", pages='all')

df = pdf[0]

df 

```

```python
# Rename column1 to the concatenation of the first and the second rows 
# in the first column

df.rename(columns={ df.columns[0]: df.iloc[0][0] + df.iloc[1][0] }, inplace = True)
# df = df.drop([0,1])

df = df.reset_index(drop=True)

df

```

```python
# Rename column2 to the value in the first row and second row  in the second column
df.rename(columns={ df.columns[1]: df.iloc[0][1] + df.iloc[1][1] }, inplace = True)



# Rename column3 to the value in the first and second row in the third column
df.rename(columns={ df.columns[2]: df.iloc[0][2] + df.iloc[1][2] }, inplace = True)


df = df.reset_index(drop=True)


df 

```

```python
# Rename column4 to the value in the first and second row in the fourth column

df.rename(columns={ df.columns[3]: df.iloc[1][3] + ' - ' + df.iloc[2][3] }, inplace = True)
df.rename(columns={ df.columns[4]: df.iloc[1][4] + ' - ' + df.iloc[2][4] }, inplace = True)
df.rename(columns={ df.columns[5]: df.iloc[1][5] + ' - ' + df.iloc[2][5] }, inplace = True)

df = df.reset_index(drop=True)

df 
```

```python
# Delete the first 3 rows 
df = df.drop([0,1,2])

df.reset_index(drop=True)

df 
```

```python
# Split column 4 
c4 = df['Decreased Eliminate - Accel Stops'].str.split('%', expand=True)
c4 


# Add a new column - Decreased - Accel
df['Decreased - Accel'] = c4[0]


# Add a new column - Eliminate - Stops 
df['Eliminate - Stops'] = c4[1]


df 


# df 
```

```python
# Remove the column 4
df = df.drop(['Decreased Eliminate - Accel Stops'], axis=1)

df 
```

Let's try another file

Extracting tables from PDFs is hard. Tabula is a tool that tries to make it easier. It works well with tables that have clearly defined borders, but it can get confused pretty easily. It's also not great at extracting text. 


Let's extract some text from a document using `PyPDF2`. We'll use the first page of a PDF file stored in the file object `pdf_file`. We'll then print the first 500 characters in the extracted text.

```python

import PyPDF2
pdf_file = open("attachment.pdf", 'rb')
# Read the pdf file
read_pdf = PyPDF2.PdfFileReader(pdf_file)
# Get the number of pages
page = read_pdf.getPage(0)
# Extract the text from the page
page_content = page.extractText()
# Print the content of the page
print(page_content[200:500])
```
