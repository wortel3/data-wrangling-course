### Structured data
Structured data are, essentially data that is organized in a tabular format. This means that the data is organized in rows and columns. Each row represents a single record, and each column represents a field in the record. Structured data is often stored in a relational database, such as MySQL, PostgreSQL, or Oracle. Structured data can also be stored in a spreadsheet, such as Microsoft Excel or Google Sheets. Structured data can also be stored in a comma-separated value (CSV) file.

```python
# !pip install openpyxl

import openpyxl

def get_number_of_sheets(file_path):
    # Load the Excel file
    workbook = openpyxl.load_workbook(file_path)

    # Get the list of sheet names
    sheet_names = workbook.sheetnames

    # Count the number of sheets
    num_sheets = len(sheet_names)

    return num_sheets

# Replace 'your_file.xlsx' with the actual path to your Excel file

print(f"Number of sheets in the spreadsheet: {num_sheets}")
```
